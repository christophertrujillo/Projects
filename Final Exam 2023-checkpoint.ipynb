# **Question 7.**
Let $r$ be the rank you found in 6. In the code cells below, make the *compact* SVD of $M$:

`M = (Hr * sr) @ CTr`

where `Hr` is a matrix with $r$ orthonormal columns, `sr` is a vector of length $r$, and `CTr` has $r$ orthonormal rows. Check that indeed `M=(Hr*sr)@CTr`, up to machine rounding.

````python
# Put your work in this code cell
r = np.sum(np.isclose(S, 0, atol=1e-10))

Hr = U[:, :r]
sr = S[:r]
CTr = VT[:r, :]

M_compact = (Hr * sr) @ CTr

print(np.allclose(M, M_compact))
````

# **Question 15.**

Make a matrix $A$ whose zero column is all $1$s and whose one column is `X`. Then, make `H,s,CT` the SVD of $A$. Use `full_matrices=False`.

````python
# Use this code cell for your answer
import numpy as np

X_reshaped = X.reshape(-1, 1)
A = np.hstack((np.ones((X.shape[0], 1)), X_reshaped))
H, s, CT = np.linalg.svd(A, full_matrices=False)

print(H)
print(s)
print(CT)
````

#### **Leave for grader.**

# **Question 16.**

Find the 2-dimensional vector $v$ that makes $A@v$ as close as possible to the vector `salaries`. Store the entries of $v$ in $b$ and $m$.

````python
# Use this code cell for your answer

v = CT.T[:, :2] @ np.diag(1/s[:2]) @ H.T[:2, :] @ salary

b,m = v
````

#### **Leave for grader**

Run this cell to print the coefficient $m$ of 'PhD Years' and the constant term $b$ from your regression.

````python
# The right hand side variables for our regression from the salary data set are listed in the
# list "variables"
# X lists just those variables from the data set

variables = ['PhD Years']

# In our regression we also need a constant term: so we make a matrix that has X as its columns, 
# and then add one more column consisting entirely of 1s.

allvariables = ['Constant Term']+variables

for item,coeff in zip(allvariables,v):
    print(f"The coefficient of {item} is {coeff:0.2f}")
````
